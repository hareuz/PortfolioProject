To get every single ticker in the table. WE first find how many items in the table, so we can iterate and get the whole list of companies. last is 101 in the page, so:

```python
tickerpath = """//*[@id="sortTableM"]/div/table[1]/tbody/tr[41]/td[1]"""
ticker = driver.find_element(by=By.XPATH, value=tickerpath)
ticker.text
```

```python
relationshiplist = []
for i in range(2,102):    
relationshippath = f"""//*[@id="sortTableM"]/div/table[1]/tbody/tr[{i}]/td[4]"""
relationship = driver.find_element(by=By.XPATH, value=relationshippath)
relationshiplist.append(relationship.text)
    
relationshippath = f"""//*[@id="sortTableM"]/div/table[1]/tbody/tr[{i}]/td[4]"""
relationship = driver.find_element(by=By.XPATH, value=relationshippath)
relationshiplist.append(relationship.text)

relationshiplist
```

```python
tickerlist = []
relationshiplist = []
datelist = []
costlist = []
shareslist = []
shareheldlist = []

for i in range(2,102):
    tickerpath = f"""//*[@id="sortTableM"]/div/table[1]/tbody/tr[{i}]/td[1]"""
    ticker = driver.find_element(by=By.XPATH, value=tickerpath)
    tickerlist.append(ticker.text)

    relationshippath = f"""//*[@id="sortTableM"]/div/table[1]/tbody/tr[{i}]/td[4]"""
    relationship = driver.find_element(by=By.XPATH, value=relationshippath)
    relationshiplist.append(relationship.text)

    datepath = f"""//*[@id="sortTableM"]/div/table[1]/tbody/tr[{i}]/td[5]"""
    date = driver.find_element(by=By.XPATH, value=datepath)
    datelist.append(date.text)

    costpath = f"""//*[@id="sortTableM"]/div/table[1]/tbody/tr[{i}]/td[6]"""
    cost = driver.find_element(by=By.XPATH, value=costpath)
    costlist.append(cost.text)

    sharespath = f"""//*[@id="sortTableM"]/div/table[1]/tbody/tr[{i}]/td[7]"""
    shares = driver.find_element(by=By.XPATH, value=sharespath)
    shareslist.append(shares.text)

    shareheldpath = f"""//*[@id="sortTableM"]/div/table[1]/tbody/tr[{i}]/td[9]"""
    shareheld = driver.find_element(by=By.XPATH, value=shareheldpath)
    shareheldlist.append(shareheld.text)
```

```python
all_info = list(zip(tickerlist, relationshiplist,datelist,costlist, shareslist, shareheldlist))
```

```python
df = pd.DataFrame(all_info, columns= ["Comoany Symbol", "Relationship", "Date","Cost", "Shares sold", "Shares held"])
df
```
